Question 1: -Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word

```python
def find_highest_frequency_highest_length_word(string):
    string = string.lower()

    words = string.split()

    word_counts = {}

    for word in words:
        if word in word_counts:
            word_counts[word]+=1
        else:
            word_counts[word] = 1

    max_frequency = max(word_counts.values())

    highest_frequet_highest_length_string = max(len(word) for word, count in word_counts.items() if count == max_frequency)
    
    return highest_frequet_highest_length_string

string_1 = "My name is amin my name is amin sharif my name"

print(find_highest_frequency_highest_length_word(string_1))

string_2 = 'There are are are most frequency frequency frequency word is is is name'
print(find_highest_frequency_highest_length_word(string_2))
```

def isvalidstring(string):
    char_count = Counter(string)
    frequencies = list(char_count.values())

    max_freq = max(frequencies)
    min_freq = min(frequencies)

    num_freq_def = sum(fre != max_freq for fre in frequencies)

    if num_freq_def == 0:
        print('YES')

    elif num_freq_def ==1 and (min_freq ==1 or min_freq == max_freq-1):
        print('YES')
    else:
        print ("No")

Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.

```python
import nltk
#nltk.download()
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag

def count_pos_tags(text):
    # Tokenize the text into sentences
    sentences = sent_tokenize(text)
    
    # Initialize count dictionary
    count_dict = {'Noun': 0, 'Verb': 0, 'Pronoun': 0, 'Adjective': 0}
    
    # Iterate over sentences
    for sentence in sentences:
        # Tokenize the sentence into words
        words = word_tokenize(sentence)
        
        # Perform part-of-speech tagging
        tagged_words = pos_tag(words)
        
        # Count the part-of-speech
        for word, tag in tagged_words:
            if tag.startswith('NN'):  # Noun
                count_dict['Noun'] += 1
            elif tag.startswith('VB'):  # Verb
                count_dict['Verb'] += 1
            elif tag.startswith('PRP'):  # Pronoun
                count_dict['Pronoun'] += 1
            elif tag.startswith('JJ'):  # Adjective
                count_dict['Adjective'] += 1
    
    return count_dict

# Example usage
text = "I love eating pizza. It is delicious and satisfying."
counts = count_pos_tags(text)
print(counts)
```
