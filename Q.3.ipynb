```python
cat_df = pd.DataFrame(df1['category'].value_counts()).reset_index()
cat_df
```

```python
cat_df.rename(columns ={'index':'news_classes', 'category':'numcat'}, inplace = True)
```

```python
cat_df
```

```python
plt.figure(figsize=(10,6))
ax = sn.barplot(np.array(cat_df.news_classes)[:10], np.array(cat_df.numcat)[:10])
plt.xticks(rotation = 90)
for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x()+0.01, p.get_height() + 150))
```

```python
df2 = df1.copy()
```

```python
df2['length_of_news'] = df2['headline'] +  df2['short_description']
df2.drop(['headline', 'short_description'], inplace = True, axis =1)
```

```python
df2['len_news'] = df2['length_of_news'].map(lambda x: len(x))
```

```python
df2
```

```python
lenmax = df2.groupby('category')['len_news'].max().reset_index().sort_values(by = 'len_news', ascending=False)
```

```python
lenmax
```

```python
lenmin = df2.groupby('category')['len_news'].min().reset_index().sort_values(by = 'len_news', ascending = False)
lenmin
```

```python
plt.figure(figsize = (15,6))

plt.subplot(121)
plt.bar(lenmax['category'][:10], lenmax['len_news'][:10])
plt.xticks(rotation = 45)
plt.ylabel('Word length in news category', size = 14)
plt.xlabel('News category', size =14)
#plt.text(6.5, 1400, ''' POLITICS HAS THE LONGEST ARTICEL''', fontsize = 12)
plt.title('Max length top 10 news categories', size = 14)

plt.subplot(122)
plt.bar(lenmin['category'][:10], lenmin['len_news'][:10])
plt.xticks(rotation = 45)
plt.ylabel('Word length in news category', size = 14)
plt.xlabel('News category', size =14)
#plt.text(6.5, 1400, ''' POLITICS HAS THE LONGEST ARTICEL''', fontsize = 12)
plt.title('Min length top 10 news categories', size = 14)
plt.show()
```
